## 1. Variable Definition

### Explanatory Variable: `Sleep`
- **What it measures:** Average hours of sleep on school nights.
- **Valid codes used:** `1–7`
- Values outside this range are treated as invalid.

### Response Variable: `SadOrHopeless`
- **What it measures:** Whether the student felt sad or hopeless.
- **Valid codes used:** `1`, `2`
- **1:** Yes
- **2:** No
- Other values are treated as invalid.

## Step 1: Check missing / invalid values

```python
sleep_col = 'Sleep'
sad_col = 'SadOrHopeless'

missing_summary = pd.DataFrame({
    'Variable':[sleep_col, sad_col],
    'Missing_Count':[
        df[sleep_col].isna().sum(),
        df[sad_col].isna().sum()
    ]
})

display(missing_summary)
```

## Step 2: Verify valid codes

```python
print("Sleep values:")
print(sorted(df['Sleep'].unique()))

print("\nSadOrHopeless values:")
print(sorted(df['SadOrHopeless'].unique()))
```

## Step 3: Final sample size

```python
final_n = len(df)

summary = pd.DataFrame({
    'Final Sample Size':[final_n],
    'Mean Sleep':[df['Sleep'].mean()],
    'SadOrHopeless Rate (%)':[
        (df['SadOrHopeless']==1).mean()*100
    ]
})

display(summary)
```

## Step 4: Save summary

```python
summary.to_csv('sleep_sad_summary.csv', index=False)
print('Summary file saved.')
```
