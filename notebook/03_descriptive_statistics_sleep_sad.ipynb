# 03 Descriptive Statistics
## Research Topic: Sleep and SadOrHopeless

Research Question:
> Is there a difference in sleep duration between students who felt sad or hopeless and those who did not?


## 1. Load Libraries & Data

```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv('YRBS_2007_sleep_sad_clean.csv')

print('Shape:', df.shape)
display(df.head())
```

## 2. Variable Exploration

```
print('Sleep Distribution')
print(df['Sleep'].value_counts().sort_index())

print('\nSadOrHopeless Distribution')
print(df['SadOrHopeless'].value_counts().sort_index())
```

## 3. Recoding

SadOrHopeless:
- 1 = Yes (felt sad or hopeless)
- 2 = No


```
df_clean = df.copy()

df_clean['sad_label'] = np.where(
    df_clean['SadOrHopeless'] == 1,
    'Yes',
    'No'
)

display(df_clean.head())
```

## 4. Descriptive Summary Table

```
summary = (
    df_clean.groupby('sad_label')['Sleep']
    .agg(
        n='count',
        mean='mean',
        median='median',
        std='std',
        minimum='min',
        maximum='max'
    )
    .round(2)
)

print('=== Descriptive Summary ===')
display(summary)
```

## 5. Visualization

```
plt.figure(figsize=(7,5))

means = summary['mean']
groups = summary.index

plt.bar(groups, means)

plt.ylabel('Average Sleep')
plt.xlabel('SadOrHopeless')
plt.title('Average Sleep by SadOrHopeless Status')

for i, v in enumerate(means):
    plt.text(i, v + 0.05, f'{v:.2f}', ha='center')

plt.show()
```

### Boxplot

```
plt.figure(figsize=(7,5))

df_clean.boxplot(
    column='Sleep',
    by='sad_label'
)

plt.title('Sleep Distribution by SadOrHopeless')
plt.suptitle('')
plt.xlabel('SadOrHopeless')
plt.ylabel('Sleep')

plt.show()
```

## 6. Save Summary Table

```
summary.to_csv(
    'descriptive_summary_sleep_by_sad.csv'
)

print('Summary table saved.')
```

## Interpretation

The summary table and figures describe the sleep patterns of students who reported feeling sad or hopeless and those who did not.

These results provide the descriptive foundation for the hypothesis testing notebook.
