## 1. Research Question

### Two-Proportion z-test

Is the proportion of students who feel sad or hopeless different between current drinkers and non-drinkers?

**中文：**  
YRBS 2007 資料中，目前有飲酒習慣與沒有飲酒習慣的高中生，感到憂鬱或絕望的比例是否有所不同？

| 項目 | 內容 |
|------|------|
| Group variable（分組變數） | `CurrentAlcoholUse`（重編碼為 binary：0 = 非飲酒者，1 = 飲酒者） |
| Response variable（反應變數） | `SadOrHopeless`（重編碼為 binary：0 = 否，1 = 是） |
| Method（方法） | Two-proportion z-test |

**假設：**

- H₀：p_drinker = p_nondrinker（飲酒者與非飲酒者感到憂鬱或絕望的比例相同）  
- H₁：p_drinker ≠ p_nondrinker（飲酒者與非飲酒者感到憂鬱或絕望的比例不同）

---

### Two-Sample t-test

Is the mean alcohol use frequency different between students who feel sad or hopeless and those who do not?

**中文：**  
YRBS 2007 資料中，感到憂鬱或絕望與未感到憂鬱或絕望的高中生，其飲酒頻率平均分數是否有所不同？

| 項目 | 內容 |
|------|------|
| Group variable（分組變數） | `SadOrHopeless`（重編碼為 binary：0 = 否，1 = 是） |
| Response variable（反應變數） | `CurrentAlcoholUse`（原始 1–7 連續值） |
| Method（方法） | Welch's Two-sample t-test |

**假設：**

- H₀：μ_sad = μ_notsad（兩組飲酒頻率平均值相同）  
- H₁：μ_sad ≠ μ_notsad（兩組飲酒頻率平均值不同）

```python
# 匯入套件
import pandas as pd

# 設定顯示格式
pd.set_option("display.max_columns", None)
pd.set_option("display.width", 120)

PROJECT = r'C:\Users\88690\OneDrive\桌面\project-cycle-3-main'

# 讀取原始資料
df = pd.read_csv(f'{PROJECT}/data/raw/YRBS_2007.csv')

# 顯示原始資料基本資訊
print('原始資料維度：', df.shape)
print('\n原始欄位名稱：')
print(df.columns.tolist())
```

```python
# 指定本專題要使用的兩個變數
selected_columns = [
    'CurrentAlcoholUse',
    'SadOrHopeless'
]

# 擷取指定欄位
df_selected = df[selected_columns].copy()

# 顯示前幾列
print('擷取後資料維度：', df_selected.shape)
print('\n前 5 列資料：')
display(df_selected.head())
```

```python
# 查看變數原始編碼分佈
print('=== CurrentAlcoholUse 原始編碼 ===')
print(df_selected['CurrentAlcoholUse'].value_counts(dropna=False).sort_index())

print()
print('=== SadOrHopeless 原始編碼 ===')
print(df_selected['SadOrHopeless'].value_counts(dropna=False).sort_index())
```

```python
# 建立輸出路徑並輸出只含兩個變數的新資料表
output_path = f'{PROJECT}/data/processed/cycle3_alcohol_sad_variables.csv'
df_selected.to_csv(output_path, index=False)

print(f'已輸出檔案：{output_path}')

# 再次確認輸出的資料表欄位
print('\n輸出資料表欄位：')
print(df_selected.columns.tolist())
```
